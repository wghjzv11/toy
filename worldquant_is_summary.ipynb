```python
dates = pd.date_range("2025-03-18", periods=3)

stock1 = pd.Series([-0.12, 0.05, -0.22], index=dates, name='stock1')
stock2 = pd.Series([-0.1, -0.1, -0.16], index=dates, name='stock2')
```

```python
table = pd.concat([stock1, stock2], axis=1).T
```

```python
table
```

```python
def linear_decay(series, window):
    weights = np.arange(1, window + 1)
    return series.rolling(window, min_periods=1).apply(
        lambda x: np.dot(x, weights[-len(x):] / weights[-len(x):].sum()),
        raw = True)
table.apply(lambda col: linear_decay(col, window=3), axis=1)
```

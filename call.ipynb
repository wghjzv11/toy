```python
import math
import numpy as np
```

```python
S0 = 100.
K = 105.
T = 1.0
r = 0.05
sigma = 0.2
```

```python
I = 100_000
```

```python
np.random.seed(1000)
```

```python
z = np.random.standard_normal(I)
```

```python
ST = S0 * np.exp((r-sigma ** 2 / 2) * T + sigma * math.sqrt(T) * z)
```

```python
hT = np.maximum(ST-K, 0)
```

```python
C0 = math.exp(-r * T) * np.mean(hT)
```

```python
print("Value of the European call option: {:5.3f}.".format(C0))
```

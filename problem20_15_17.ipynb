```python
def put(S: float,
        K: float,
        r: float,
        T: float,
        call_price: float):
    return call_price + K * math.exp(-r * T) - S
```

```python
put(S=40, K=50, r=0.05, T=0.5, call_price=0.725)
```

```python
put(S=40, K=50, r=0.05, T=0.5, call_price=11.15)
```

```python
quotes_call = [
    (42, 8.42),
    (44, 7.37),
    (46, 6.31),
    (48, 5.26),
    (50, 4.21),
    (52, 3.16),
    (54, 2.10),
    (56, 1.05),
    (58, 0.00)
]

iv_call = pd.Series(
    {K: bsm.get_implied_vol(price=price, 
                            S=50,
                            K=K,
                            r=0.12,
                            T=1/12,
                            option="call",
                            q=0.0) for K, price in quotes_call},
    name="implied_vol"
).sort_index()
```

```python
print(iv_call)
```

```python
quotes_put = [
    (42, 0),
    (44, 0.93),
    (46, 1.86),
    (48, 2.78),
    (50, 3.71),
    (52, 4.64),
    (54, 5.57),
    (56, 6.50),
    (58, 7.42)
]

iv_put = pd.Series(
    {K: bsm.get_implied_vol(price=price, 
                            S=50,
                            K=K,
                            r=0.12,
                            T=1/12,
                            option="put",
                            q=0.0) for K, price in quotes_put},
    name="implied_vol"
).sort_index()
```

```python
iv_put
```

```python
# 1. 1행 2열 구조의 서브플롯 생성
# figsize는 전체 그림의 크기(가로, 세로) 인치 단위입니다.
# 1. 1행 2열 구조 생성
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 2. 첫 번째 그래프 (왼쪽): iv.plot을 사용
# ax=axes[0] 라고 지정하면 해당 칸에 그래프가 그려집니다.
iv_call.plot(marker="o", ax=axes[0]) 
axes[0].set_xlabel("Strike price")
axes[0].set_ylabel("Implied Volatility")
axes[0].set_title("Volatility Smile(call)")

iv_put.plot(marker="o", ax=axes[1])
axes[1].set_xlabel("Strike price")
axes[1].set_ylabel("Implied Volatility(put)")
axes[1].set_title("Volatility Smile(put)")
```

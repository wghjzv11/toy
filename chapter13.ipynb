```python
# 13.21
p_star = 0.6266
call_option_expected_value = p_star * 1
print(call_option_expected_value)
call_option_value = binomial.simple_binomial_tree(
    S=20,
    K=21,
    r=0.04,
    T=3/12,
    up=22,
    down=18,
    option_type="call"
)
discount_rate = np.log(call_option_value/call_option_expected_value)*(-4)
print(f"discount rate:{discount_rate}")
```

```python
put_option_expected_value = (1-p_star) * 3
put_option_value = binomial.simple_binomial_tree(
    S=20,
    K=21,
    r=0.04,
    T=3/12,
    up=22,
    down=18,
    option_type="put"
)
discount_rate = np.log(put_option_value/put_option_expected_value)*(-4)
print(f"discount rate:{discount_rate}")
```

해설지는 CAPM 모델의 가설로 설명하고 있다.
콜옵션은 시장 수익률과 같은 방향이므로 더 많은 프리미엄을 요구하고,
풋옵션은 시장 수익률의 반대 방향이므로 더 적은 프리미엄을 요구한다는 것이다.

그러나, 생각해 보면 binomial model 자체가 capm을 채택하고 있는지는 고려해봐야 하는 것 같다.

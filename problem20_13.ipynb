```python
import bsm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

![problem20_13](images/problem20_13.png)

```python
# Risk-neutral probability
p = 0.4
```

```python
help(bsm.get_price)
```

```python
# Situation 1. Ruling was Favorable.
Ks = np.linspace(30, 80, 21)
upward = pd.Series({k: bsm.get_price(
              S=75.0,
              K=k,
              r=0.06,
              T=0.5,
              vol=0.25,
              option="call",
              q=0.0) for k in Ks})

downward = pd.Series({k: bsm.get_price(
              S=50.0,
              K=k,
              r=0.06,
              T=0.5,
              vol=0.40,
              option="call",
              q=0.0) for k in Ks})
              
```

```python
prices = p * upward + (1-p) * downward
```

```python
prices
```

```python
help(bsm.get_implied_vol)
```

```python
iv = pd.Series({k: bsm.get_implied_vol(
    price = price,
    S=60,
    K=k,
    r=0.06,
    T=0.5,
    option="call",
    q=0.0
) for k, price in prices.items()})
```

```python
iv
```

```python
iv.plot(marker="o")
plt.title("Volatility Smile Analysis", fontsize=15, fontweight='bold')
plt.xlabel("Strike Price (K)")
plt.ylabel("Implied Volatility (Ïƒ)")
```
